```python
def phs(b):
    if (b >= 0.0) & (b < 36.0):
        return 'pp'
    elif (b >= 36.0) & (b < 96.0):
        return 'mid'
    else:
        return 'death'
```

```python
balls = 114
wkt = 5
for pl in dictteams['CSK'][wkt+2:]:
    ph = phs(balls)     
    print(ph)
    print(playersData[pl]['bat'][ph]['wkt'], pl)
    balls += 1/playersData[pl]['bat'][ph]['wkt']
    print(balls)
    print()
agg = balls/(120 + (10 -wkt-2)*3)
agg = balls/140
agg
```

```python
adjustedDataAll = {}
adjustedDataAll['wkt'] = playersData[playerName]['bat'][phase]['wkt'] * agg
adjustedDataAll['0'] = playersData[playerName]['bat'][phase]['0'] * 1/agg
adjustedDataAll['1'] = playersData[playerName]['bat'][phase]['1'] * 1/agg
adjustedDataAll['2'] = playersData[playerName]['bat'][phase]['2'] * agg
adjustedDataAll['3'] = playersData[playerName]['bat'][phase]['3'] * agg
adjustedDataAll['4'] = playersData[playerName]['bat'][phase]['4'] * agg
adjustedDataAll['6'] = playersData[playerName]['bat'][phase]['6'] * agg
```

```python
 adjustedDataAll['sr'] = round((1*adjustedDataAll['1'] + 2*adjustedDataAll['2'] + 3*adjustedDataAll['3'] + 4*adjustedDataAll['4'] + 6*adjustedDataAll['6']) / (adjustedDataAll['wkt'] + adjustedDataAll['0'] + adjustedDataAll['1'] + 
 adjustedDataAll['2'] + adjustedDataAll['3'] + 
 adjustedDataAll['4'] + adjustedDataAll['6']), 4)
```
