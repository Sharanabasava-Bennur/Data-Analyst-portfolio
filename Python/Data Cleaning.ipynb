```python
import pandas as pd
from copy import deepcopy
import numpy as np
```

```python
path = r'.\Datasets\Customer Call List.xlsx'
```

```python
dfRaw = pd.read_excel(path)
```

```python
df = dfRaw.copy(deep='True')
```

```python
df
```

```python
df.drop('Not_Useful_Column', axis=1, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df['Paying Customer'].value_counts()
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('123./')
```

```python
convert_dict = {'Phone_Number': str
                }
 
df = df.astype(convert_dict)
```

```python
df.replace({'Phone_Number' : '[-/|]'}, {'Phone_Number' : ''}, regex=True, inplace=True)
```

```python
# df["Phone_Number"] = df["Phone_Number"].str.replace('NaN','')

# df["Phone_Number"] = df["Phone_Number"].str.replace('Na','')
# df.replace({}np.nan, '', regex=True, inplace=True)
```

```python
df.replace({'Phone_Number' : [np.nan, 'Na', 'nan']}, {'Phone_Number' : ''}, inplace=True)
```

```python
df['Phone_Number'] = df["Phone_Number"].apply(lambda x : x[0:3]+'-'+x[3:6]+'-'+x[6:10] if len(x) else '')
```

```python
df['Phone_Number']
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2, expand=True)
```

```python
df
```

```python
df['Paying Customer'].value_counts()
```

```python
df.replace({'Paying Customer' : ['Yes', 'No']}, {'Paying Customer' : ['Y', 'N']}, inplace=True)
```

```python
df['Do_Not_Contact'].value_counts()
```

```python
df.replace({'Do_Not_Contact' : ['Yes', 'No']}, {'Do_Not_Contact' : ['Y', 'N']}, inplace=True)
```

```python
df.dropna(subset=['Phone_Number'])
```

```python
# df.fillna('')
df.replace(['N/a', np.nan], '', inplace=True)
```

```python
finalisedCallList = df[(df['Phone_Number']!='') & (df['Do_Not_Contact']!='Y')]
```

```python
pd.options.mode.copy_on_write = True
```

```python
finalisedCallList.reset_index(drop=True, inplace=True)
```

```python
finalisedCallList
```

```python
finalisedCallList.to_excel(r'.\Datasets\Cleaned Customer Call List.xlsx')
```
